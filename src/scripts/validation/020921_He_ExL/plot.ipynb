```python
import matplotlib.pyplot as plt

import pandas as pd
import pickle as pk

def load_pk(path):
    with open(path, 'rb') as f:
        x = pk.load(f)
    return x

def decorate(ax, title=None, xaxis=None, yaxis=None, xlims=None, ylims=None):
    ax.set_title(title)
    ax.set_xlabel(xaxis)
    ax.set_ylabel(yaxis)
    if xlims is not None: ax.set_xlim(xlims) 
    if ylims is not None: ax.set_ylim(ylims)


```

```python

isolated_energy = -7.477

fig, axs = plt.subplots(1, 1, figsize=(6, 6))  # x, y
if type(axs) is list:
    axs = [ax for lst in axs for ax in lst]

path = 'results2.csv'
# data = load_pk(path)
data = pd.read_csv(path, header=None).sort_values(by=0)

X = data.iloc[:, 0]
Y = data.iloc[:, 1]

axs.plot(X, Y)
axs.plot(X, [isolated_energy for _ in X], color='r', ls='--')

decorate(axs, 'Li single nuclei in cell', 'Cell width (A)', 'Energy', ylims = (-10, 30))

fig.tight_layout()

plt.show()
```
