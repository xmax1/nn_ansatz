```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.cm import tab10
tab10 = iter(tab10)
```

```python
x = np.linspace(-1, 1, 100)

def cos(x, n=1):
    return np.cos(2 * np.pi * n * x)

def sin(x, n=1):
    return np.sin(2 * np.pi * n * x)

n_p = 3
ins = []
for i in range(1, n_p):
    # ins.extend([cos(x, n=i), sin(x, n=i)])
    ins.extend([cos(x, n=i)])

plt.rcParams.update({
        "text.usetex": True,
        "font.family": "Times New Roman"})

latex = True
title = ''
xaxis = r'$\boldmath{r}$'
yaxis = r'$p( \boldmath{r})$'
xlims = (-1, 1)
ylims = (-1.1, 1.1)
xlines = [-0.5, 0.5]

fig, ax = plt.subplots(1, 1, figsize=(4, 4))
ax.set_title(title)
ax.set_xlabel(xaxis)
ax.set_ylabel(yaxis)
if xlims is not None: ax.set_xlim(xlims) 
if ylims is not None: ax.set_ylim(ylims)

ax.set_xticks([])
ax.set_yticks([])

ax.grid(b=False, which='major', axis='both')
[ax.axvline(x=x, color='red', ls='--', alpha=0.5) for x in xlines]

colors = [0, 2]
for i, f in zip(colors, ins):
    plt.plot(x, f, color=tab10(i))

fig.savefig('pin_svg.svg', format='svg', dpi=1200)

plt.show()

    
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
n = 100
s = 1/2
x_lim = 30
x = np.outer(np.linspace(-x_lim, x_lim, n), np.ones(n))
y = x.copy().T # transpose
z = np.abs(np.cos(s*x)  +  np.cos(s*y) ) / 10

fig = plt.figure(figsize=(15, 15))
ax = plt.axes(projection='3d')
# Make panes transparent
ax.xaxis.pane.fill = False # Left pane
ax.yaxis.pane.fill = False # Right pane

# Remove grid lines
ax.grid(False)
plt.axis('off')

# Remove tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

ax.set_zlim(0,2)

ax.view_init(azim=15, elev=10)
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none', rstride=1, cstride=1, shade=True, linewidth=0)
# ax.set_title('Surface plot')

fig.tight_layout(pad=0.)
plt.show()
```
